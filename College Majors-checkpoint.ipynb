## Histograms

```python
recent_grads['Sample_size'].hist(bins = 16, range= (0,800),grid=False)
print(recent_grads['Sample_size'].describe())
```

```python
recent_grads['Sample_size_pctg'] = (recent_grads['Sample_size'])/(recent_grads['Total'])*100
print(recent_grads['Sample_size_pctg'].describe())
```

There is an easily visible Sample Size curve. With the most frequency being between 0 to 50 people sampled.

The **sample/total people enrolled** formula is high, which means the data was well gathered with good samples.

```python
recent_grads['Median'].hist(bins= 20, grid=False, range= (20000,80000))
```

Most new-grads earn between 30,000 to 40,000USD

```python
recent_grads['Employed_pctg'] = (recent_grads['Employed'])/(recent_grads['Total'])*(100)
```

```python
recent_grads['Employed_pctg'].hist(bins= 20, grid=False, range= (0,100))
```

Most majors have employment rates from 70% to 90% 

```python
recent_grads['Full_time_pctg'] = (recent_grads['Full_time'])/(recent_grads['Total'])*(100)
recent_grads['Full_time_pctg'].hist(bins= 20, grid=False, range=(0,100))
```

60 to 80 % are in full time jobs 

```python
recent_grads['ShareWomen'].hist(bins= 5, grid=False)
```

48% of Majors have mor men than women

```python
recent_grads['Unemployment_rate'].hist(bins= 10, grid=False)
```

Unemployment mainly ranges from 2.5% to 12.5%
